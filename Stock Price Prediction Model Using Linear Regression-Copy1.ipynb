```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn import linear_model 
```

```python
df= pd.read_csv("C:/Users/jadha/Downloads/Quote-Equity-LT-EQ-01-07-2023-to-01-08-2023.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y')
df=df.sort_values(by='Date',ascending=True)
df
```

```python
numeric_columns = ['OPEN', 'HIGH', 'LOW', 'PREV. CLOSE', 'close', 'VOLUME', 'Next_Open', 'No of trades']

# Shift 'OPEN' column to 'Next_Open'
df['Next_Open'] = df['OPEN'].shift(-1)

# Drop rows with missing values
df.dropna(inplace=True)

# Convert numeric columns to numeric type
for col in numeric_columns:
    df[col] = df[col].str.replace(',', '').astype(float)

df
```

```python

    
input_columns = ['OPEN', 'HIGH', 'LOW', 'PREV. CLOSE','close','Next_Open' ]
output_columns = ['HIGH', 'LOW',]
   

input_val = df[input_columns].iloc[:-1]
output_val = df[output_columns].iloc[1:]

input_val
```

```python
# Create and train your regression model
reg = linear_model.LinearRegression()
reg.fit(input_val, output_val)
```

```python

df.drop(df.index[-1], inplace=True)
df
```

```python

# Now you can use the trained model for prediction
dopen = float(input("entrer open value:" ))
dhigh = float(input("entrer high value:" ))
dlow = float(input("entrer low value:" ))
dprevclose = float(input("entrer prevclose value:" ))
dclose = float(input("entrer close value:" ))
dnxtopen = float(input("entrer nxtopen value:" ))
input_values = [[dopen, dhigh, dlow, dprevclose,dclose,dnxtopen]]
predicted_values = reg.predict(input_values)
print(predicted_values)

```
